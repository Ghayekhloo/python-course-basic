## 1. Why Do We Need Functions?
Functions help reuse code.


```python
def say_hello():
    print("Hello student!")

say_hello()
```

## 2. Defining and Calling a Function

```python
def greet():
    print("Welcome to Python")

greet()
```

## 3. Functions with Parameters

```python
def student_info(name, major):
    print(f"{name} studies {major}")

student_info("Lina", "Marine ecology")
```

## 4. Functions That Return Values

```python
def add(a,b):
    return a+b

print(add(2,3))
```

```python
def classify(score):
    if score>=85:
        return "Excellent"
    elif score>=70:
        return "Good"
    else:
        return "Needs improvement"

print(classify(78))
```

## 5. Functions + Lists + Loops

```python
def avg_list(lst):
    return sum(lst)/len(lst)

print(avg_list([80,90,100]))
```

## 6. Input + Function

```python
def bmi(w,h):
    return w/(h**2)

print(bmi(70,1.75))
```

## 7. Helper Functions

```python
def major_advice(major):
    if major=="Marine ecology":
        return "Ocean studies"
    elif major=="Mechanical Engineering":
        return "Machines and design"
    else:
        return "General field"

print(major_advice("Mechanical Engineering"))
```

## Final Example

```python
def advisor(name,major,gpa):
    if gpa>=3.5:
        level="Excellent"
    elif gpa>=3.0:
        level="Good"
    else:
        level="Weak"
    return f"{name}-{major}-{level}"

print(advisor("Sara","Communication Design",3.6))
```
