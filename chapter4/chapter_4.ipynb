## 1. Why do we use functions?

Without functions, we repeat the same code many times. With functions:
- we write code **once**
- we call it many times
- our programs become cleaner and easier to read.

## 2. First simple functions (no parameters)

Syntax:
```python
def function_name():
    # body (indented)
```

````python
def say_hello():
    print("Hello, welcome to the Python class!")

# call the function several times
say_hello()
say_hello()
````

````python
def print_separator():
    print("---------------------")

print("Start of program")
print_separator()
print("Some important results")
print_separator()
print("End of program")
````

## 3. Functions with parameters

Parameters are variables in the function definition.
Arguments are the actual values we pass when we call the function.

````python
def greet(name):
    print("Hello", name)

# using your style with real names
greet("Sara")
greet("Marco")
greet("Claudia")
````

````python
# function with two parameters

def print_full_name(first_name, last_name):
    print("Full name:", first_name, last_name)

print_full_name("Omar", "Ahmad")
print_full_name("Sara", "Ghayekhloo")
````

````python
# function with different data types

def print_student(name, age, is_student):
    print("Name:", name)
    print("Age:", age)
    print("Is a student?", is_student)

print_student("Claudia", 23, True)
````

## 4. Functions that RETURN a value

So far functions only printed text. Now we want functions that **calculate** and then **return** a value.

````python
def add(a, b):
    result = a + b
    return result

x = add(3, 5)
print("Result:", x)
````

````python
# average of three scores (similar to your score examples)

def average_of_three(a, b, c):
    total = a + b + c
    avg = total / 3
    return avg

score1 = 18
score2 = 25
score3 = 20

avg_score = average_of_three(score1, score2, score3)
print("Average score:", avg_score)
````

````python
# classify a score using if / elif / else inside a function

def classify_score(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    elif score >= 60:
        return "C"
    else:
        return "F"

my_score = 72
result = classify_score(my_score)
print("Score:", my_score, "-> Grade:", result)
````

## 5. Functions + lists + for loop

Now we combine functions with lists, like you did before with scores and temperatures.

````python
# sum of a list

def sum_list(numbers):
    total = 0
    for n in numbers:
        total = total + n
    return total

scores = [10, 15, 20, 25]
print("Scores:", scores)
print("Total score:", sum_list(scores))
````

````python
# average of a list

def average_list(values):
    total = 0
    for v in values:
        total += v
    return total / len(values)

temperatures = [20.5, 22.0, 19.5, 21.0]
print("Average temperature:", average_list(temperatures))
````

````python
# count how many scores are PASS (>= 18)

def count_pass(scores):
    count = 0
    for s in scores:
        if s >= 18:
            count += 1
    return count

scores = [12, 18, 25, 10, 30]
print("Scores:", scores)
print("Number of students who passed:", count_pass(scores))
````

## 6. Functions and input from the user

We can put `input()` inside a function or we can pass the input as a parameter.

````python
# example: read age from input and return it as int

def read_age():
    age_text = input("Enter your age: ")
    age = int(age_text)
    return age

user_age = read_age()
print("You are", user_age, "years old.")
````

````python
# BMI calculator (weight & height from input)

def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi

w = float(input("Enter your weight in kg: "))
h = float(input("Enter your height in meters: "))

bmi_value = calculate_bmi(w, h)
print("Your BMI is:", bmi_value)
````

## 7. Small helper functions (mini example with your style)

We create several small functions and use them together.

````python
# function to print a student line (similar to your f-string examples)

def print_student_line(name, field, age):
    print(f"My name is {name}, I study {field}, and I am {age} years old.")

print_student_line("Sara", "Economics", 27)
print_student_line("Omar", "Mechanical Engineering", 24)
````

````python
# combine: read 3 scores from user, compute average, and classify

def read_three_scores():
    s1 = float(input("Enter score 1: "))
    s2 = float(input("Enter score 2: "))
    s3 = float(input("Enter score 3: "))
    return s1, s2, s3

# reuse previous functions

def average_of_three(a, b, c):
    total = a + b + c
    return total / 3

def classify_score(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    elif score >= 60:
        return "C"
    else:
        return "F"

s1, s2, s3 = read_three_scores()
avg = average_of_three(s1, s2, s3)
letter = classify_score(avg)

print("Average:", avg)
print("Final grade:", letter)
````
