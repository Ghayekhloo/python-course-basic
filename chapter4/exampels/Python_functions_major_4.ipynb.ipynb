## 1. Why Do We Need Functions?

Functions help us:
- avoid repeating the same code
- organize code into small parts
- make programs easier to read
- solve real problems step by step

In teaching, functions are very useful because students can focus on **one idea at a time**.


```python
# Without a function: repeated code
print("Welcome, Marine ecology student!")
print("Welcome, Mechanical Engineering student!")
print("Welcome, Communication Design student!")

# With a function: reusable code
def welcome_student(major):
    print(f"Welcome, {major} student!")

welcome_student("Marine ecology")
welcome_student("Mechanical Engineering")
welcome_student("Communication Design")
```

### Why this matters for your students

Imagine that you want to write advice for many different majors:
- Supply chain management
- Marine ecology
- Electronics Engineering
- Biodiversity
- Mechanical Engineering

Instead of writing the same structure again and again, we create a function once and call it many times.


```python
def major_message(major):
    print(f"Today we are exploring Python examples for {major}.")

major_message("Supply chain management")
major_message("Marine ecology")
major_message("Electronics Engineering")
major_message("Biodiversity")
major_message("Mechanical Engineering")
```

## 3. Functions with Parameters

Parameters make functions more flexible.

A parameter is a variable inside the function definition.
When we call the function, we give it a value.


### 3.1 What Is a Parameter?

In the function below, `major` is a parameter.


```python
def show_major(major):
    print(f"Student major: {major}")

show_major("Marine ecology")
show_major("Electronics Engineering")
```

### 3.2 Example 1 – Function with one parameter

This function gives one custom message for one major.


```python
def major_focus(major):
    print(f"{major} students can use Python to solve field-specific problems.")

major_focus("Biodiversity")
major_focus("Vegetation science")
major_focus("Mechanical Engineering")
```

### 3.3 Example 2 – Function with two parameters

Now we use both `name` and `major`.


```python
def student_intro(name, major):
    print(f"{name} studies {major}.")

student_intro("Lina", "Marine ecology")
student_intro("Omar", "Mechanical Engineering")
student_intro("Nora", "Communication Design")
```

### 3.4 Example 3 – Parameters and types

A function can receive:
- strings

- booleans

This example combines student information in one function.


```python
def student_profile(name, major, gpa, has_project):
    print(f"Name: {name}")
    print(f"Major: {major}")
    print(f"GPA: {gpa}")
    print(f"Completed project: {has_project}")

student_profile("Sara", "Electronics Engineering", 3.8, True)
print()
student_profile("Daniel", "Biodiversity", 3.2, False)
```

## 4. Functions That Return a Value

Some functions do not only print a message.
They **return** a value, and we can store or use that value later.


### 4.1 Example 1 – Simple sum

```python
def add_numbers(a, b):
    return a + b

result = add_numbers(12, 8)
print("Result:", result)

# Example related to majors
lab_score = 15
quiz_score = 18
total = add_numbers(lab_score, quiz_score)
print("Total score:", total)
```

### 4.2 Example 2 – Average of three scores

This is useful for all student majors.


```python
def average_of_three(score1, score2, score3):
    return (score1 + score2 + score3) / 3

marine_avg = average_of_three(85, 91, 88)
design_avg = average_of_three(92, 87, 90)
engineering_avg = average_of_three(78, 84, 80)

print("Marine ecology average:", marine_avg)
print("Communication Design average:", design_avg)
print("Mechanical Engineering average:", engineering_avg)
```

### 4.3 Example 3 – Classifying a score

This function returns a text result.


```python
def classify_score(score):
    if score >= 90:
        return "Excellent"
    elif score >= 75:
        return "Good"
    elif score >= 60:
        return "Pass"
    else:
        return "Needs improvement"

print("Supply chain management:", classify_score(93))
print("Marine ecology:", classify_score(81))
print("Electronics Engineering:", classify_score(67))
print("Biodiversity:", classify_score(52))
```

## 5. Functions, Lists, and Loops Together

This section combines ideas from Notebook 3:
- lists
- `for` loops
- `while` loops
- functions

This is where students start writing more realistic programs.


### 5.1 Example 1 – Sum of a list

```python
def sum_of_list(numbers):
    total = 0
    for number in numbers:
        total += number
    return total

marine_scores = [18, 20, 17, 19]
electronics_scores = [15, 16, 18, 20]

print("Marine ecology total:", sum_of_list(marine_scores))
print("Electronics Engineering total:", sum_of_list(electronics_scores))
```

### 5.2 Example 2 – Average of a list

```python
def average_of_list(numbers):
    return sum(numbers) / len(numbers)

supply_chain_scores = [80, 85, 78, 90]
design_scores = [88, 92, 84, 91]
biodiversity_scores = [76, 81, 79, 85]

print("Supply chain management average:", average_of_list(supply_chain_scores))
print("Communication Design average:", average_of_list(design_scores))
print("Biodiversity average:", average_of_list(biodiversity_scores))
```

### 5.3 Example 3 – Count how many pass

This function counts scores that are 60 or higher.


```python
def count_pass(scores):
    count = 0
    for score in scores:
        if score >= 60:
            count += 1
    return count

mechanical_scores = [55, 60, 72, 81, 49]
vegetation_scores = [61, 63, 70, 58, 75]

print("Mechanical Engineering pass count:", count_pass(mechanical_scores))
print("Vegetation science pass count:", count_pass(vegetation_scores))
```

### Extra strong example – Functions + lists + loops + majors

This example gives academic advice to many students.


```python
def academic_status(gpa):
    if gpa >= 3.7:
        return "Outstanding"
    elif gpa >= 3.0:
        return "Good"
    else:
        return "Needs support"

student_names = ["Lina", "Omar", "Nora", "Daniel", "Mila"]
student_majors = [
    "Marine ecology",
    "Mechanical Engineering",
    "Communication Design",
    "Electronics Engineering",
    "Biodiversity"
]
student_gpas = [3.8, 3.1, 3.9, 2.8, 3.4]

for i in range(len(student_names)):
    name = student_names[i]
    major = student_majors[i]
    gpa = student_gpas[i]
    print(f"{name} | {major} | {academic_status(gpa)}")
```

### 6.2 Example 2 – BMI calculator

This is a common beginner example for functions with input and return values.


```python
def calculate_bmi(weight, height):
    return weight / (height ** 2)

bmi_value = calculate_bmi(70, 1.75)
print("BMI:", round(bmi_value, 2))
```

### Stronger applied example – input + classification

This example gives simple academic feedback.


```python
def student_feedback(name, major, gpa):
    if gpa >= 3.7:
        return f"{name} from {major} is an outstanding student."
    elif gpa >= 3.0:
        return f"{name} from {major} is doing well."
    else:
        return f"{name} from {major} should practice more and ask for support."

# Example:
print(student_feedback("Sara", "ecological Modeling", 3.6))
```

## 7. Designing Small Helper Functions

Helper functions are small functions that do one specific job.

They are very useful in larger programs because they keep code clean and readable.


```python
def is_science_major(major):
    science_majors = ["Marine ecology", "Biodiversity", "Vegetation science", "ecological Modeling"]
    return major in science_majors

print(is_science_major("Marine ecology"))
print(is_science_major("Mechanical Engineering"))
```

```python
def is_engineering_major(major):
    engineering_majors = [
        "Electronics Engineering",
        "Mechanical Engineering",
        "Mechanical engineer",
        "Mechanical manufacturing management and aerospace innovation"
    ]
    return major in engineering_majors

print(is_engineering_major("Electronics Engineering"))
print(is_engineering_major("Biodiversity"))
```

```python
def major_advice(major):
    if major == "Supply chain management":
        return "Focus on logistics, planning, and operations."
    elif major == "Communication Design":
        return "Build a strong portfolio and practice visual communication."
    elif major == "Marine ecology":
        return "Strengthen fieldwork, observation, and environmental analysis."
    elif major == "Electronics Engineering":
        return "Practice circuits, coding, and embedded systems."
    elif major == "Biodiversity":
        return "Develop reporting and conservation analysis skills."
    elif major == "Vegetation science":
        return "Improve plant identification and ecosystem mapping."
    elif major == "Mechanical Engineering" or major == "Mechanical engineer":
        return "Focus on mechanics, CAD, and engineering problem solving."
    elif major == "ecological Modeling":
        return "Use Python, mathematics, and simulation in ecosystem studies."
    else:
        return "Combine innovation, manufacturing, and aerospace thinking."

for major in [
    "Supply chain management",
    "Communication Design",
    "Marine ecology",
    "Electronics Engineering",
    "Biodiversity",
    "Vegetation science",
    "Mechanical Engineering",
    "ecological Modeling",
    "Mechanical manufacturing management and aerospace innovation"
]:
    print(major, "->", major_advice(major))
```

## Final Comprehensive Example

This final example combines:
- functions
- parameters
- return values
- lists
- loops
- `if / elif / else`
- helper functions
- real majors

It acts like a small student advisor system.


```python
def academic_status(gpa):
    if gpa >= 3.7:
        return "Outstanding"
    elif gpa >= 3.0:
        return "Good"
    else:
        return "Needs support"

def major_advice(major):
    if major == "Supply chain management":
        return "Improve logistics analysis and decision making."
    elif major == "Communication Design":
        return "Build visual projects and presentation skills."
    elif major == "Marine ecology":
        return "Strengthen environmental observation and research practice."
    elif major == "Electronics Engineering":
        return "Practice electronics, sensors, and programming."
    elif major == "Biodiversity":
        return "Focus on conservation, ecosystems, and reporting."
    elif major == "Vegetation science":
        return "Improve plant science and ecological mapping skills."
    elif major == "Mechanical Engineering":
        return "Develop mechanics, simulation, and design confidence."
    elif major == "ecological Modeling":
        return "Combine coding, mathematics, and ecological simulation."
    else:
        return "Work on innovation, manufacturing, and aerospace systems."

def student_advisor(name, major, gpa):
    status = academic_status(gpa)
    advice = major_advice(major)
    return f"{name} | {major} | {status} | {advice}"

student_names = ["Lina", "Omar", "Sara", "Daniel", "Nora", "Mila"]
student_majors = [
    "Marine ecology",
    "Mechanical Engineering",
    "Communication Design",
    "Electronics Engineering",
    "Biodiversity",
    "Supply chain management"
]
student_gpas = [3.8, 3.2, 3.9, 2.7, 3.5, 3.1]

for i in range(len(student_names)):
    print(student_advisor(student_names[i], student_majors[i], student_gpas[i]))
```
