## Part 3 - Final Project: Student Grade Manager

We build a real program in 5 steps. Each step is one function.

### Step 1 - Collect student names and scores

```python
def collect_students():
    names = []
    scores = []
    n = int(input('How many students? '))
    for i in range(n):
        name  = input(f'Enter name of student {i+1}: ')
        score = float(input(f'Enter score for {name}: '))
        names.append(name)
        scores.append(score)
    return names, scores

names, scores = collect_students()
print('Names:', names)
print('Scores:', scores)
```

### Step 2 - Compute the average

```python
def average(scores):
    total = 0
    for s in scores:
        total += s
    return total / len(scores)

print('Class average:', average(scores))
```

### Step 3 - Classify each score

```python
def classify(score):
    if score >= 18:
        return 'Pass'
    else:
        return 'Fail'

print(classify(20))
print(classify(12))
```

### Step 4 - Print the report

```python
def print_report(names, scores):
    print('\n--- Student Report ---')
    for i in range(len(names)):
        status = classify(scores[i])
        print(f'{names[i]:15} | Score: {scores[i]:5.1f} | {status}')
    print(f'\nClass average: {average(scores):.2f}')

print_report(names, scores)
```

### Step 5 - Save the report to a file

```python
def save_report(filename, names, scores):
    with open(filename, 'w') as f:
        f.write('Student Grade Report\n')
        f.write('====================\n')
        for i in range(len(names)):
            status = classify(scores[i])
            f.write(f'{names[i]} | {scores[i]} | {status}\n')
        f.write(f'\nClass average: {average(scores):.2f}\n')

save_report('grade_report.txt', names, scores)
print('Report saved.')

with open('grade_report.txt', 'r') as f:
    print('\n--- File Content ---')
    print(f.read())
```

### Step 6 - Main program (all together)

```python
names, scores = collect_students()
print_report(names, scores)
save_report('grade_report.txt', names, scores)
print('\nDone! Report saved to grade_report.txt')
```

## Part 4 - Extension Challenges

```python
# Challenge 1: find the student with the highest score
def top_student(names, scores):
    max_score = scores[0]
    top_name  = names[0]
    for i in range(len(scores)):
        if scores[i] > max_score:
            max_score = scores[i]
            top_name  = names[i]
    return top_name, max_score

top, s = top_student(names, scores)
print(f'Top student: {top} with score {s}')
```

```python
# Challenge 2: count passed and failed
def count_results(scores):
    passed = 0
    failed = 0
    for s in scores:
        if s >= 18:
            passed += 1
        else:
            failed += 1
    return passed, failed

passed, failed = count_results(scores)
print(f'Passed: {passed} | Failed: {failed}')
```

```python
# Challenge 3: letter grades
def letter_grade(score):
    if score >= 27:
        return 'A'
    elif score >= 24:
        return 'B'
    elif score >= 18:
        return 'C'
    else:
        return 'F'

print('\n--- Letter Grades ---')
for i in range(len(names)):
    print(f'{names[i]:15} | {scores[i]} | {letter_grade(scores[i])}')
```
