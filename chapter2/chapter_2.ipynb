## 2. Basic `if / elif / else` examples

```python
# Example 1: score evaluation
score = 20

if score > 90:
    print("Excellent")
elif score > 70:
    print("Good")
else:
    print("Need improvement")
```

```python
# Example 2: accuracy of a device
accuracy = 0.95

if accuracy > 0.9:
    print("High accuracy device")
elif accuracy > 0.7:
    print("Medium accuracy device")
else:
    print("Low accuracy device")
```

**Note:**
- `if` → first condition
- `elif` → next condition(s) if the previous ones were `False`
- `else` → everything else (when none of the above were `True`).

## 3. Input as `string`

```python
# Example 3: name from input
name = input("Enter your name: ")
print(f"Hello {name}")
print(type(name))  # always <class 'str'>
```

```python
# Example 4: age as string
age = input("Enter your age: ")
print(f"You are {age} years old")
print(type(age))  # still a string
```

`input()` **always** returns a value of type `str` (string).
If you need a number, you must convert it with `int()` or `float()`.

## 4. Input converted to `int`

```python
# Example 5: age as integer
age = int(input("Enter your age: "))

if age > 18:
    print("Adult")
else:
    print("Not adult")

print(type(age))  # now it is an int
```

```python
# Example 6: incorrect comparison (string vs string)
# Here age is NOT converted to int, so we compare strings
age = input("Enter your age again: ")

if age > "18":
    print("Adult (string comparison)")
else:
    print("Young (string comparison)")

print(type(age))  # still a string
```

In the second example, `age` is a **string`.
Python compares strings alphabetically, not numerically, so results can be confusing.
Always convert to `int` when you really mean a number.

## 5. Input converted to `float`

```python
# Example 7: water temperature classification

water_temp = float(input("Enter water temperature (°C): "))

if water_temp > 25:
    print("Warm water")
elif water_temp > 15:
    print("Moderate water")
else:
    print("Cold water")
```

## 6. Input and boolean flags

```python
# Example 8: sea water cleanliness
sea_name = input("Enter sea name: ")
# user types 'yes' if water is clean
is_water_clean = input("Is the water clean? (yes/no): ") == "yes"

status = "clean" if is_water_clean else "polluted"

print(f"In {sea_name}, the water is {status}.")
print(type(is_water_clean))  # this is a bool
```

Here we convert a text answer to a **boolean**:
- If the user types `yes` → `is_water_clean` becomes `True`
- Otherwise → it becomes `False`.

## 7. New practice examples (added)

```python
# Example 9: loan approval
income = float(input("Enter your monthly income (EUR): "))
credit_score = int(input("Enter your credit score (0-1000): "))

if income >= 3000 and credit_score >= 700:
    print("Loan approved.")
elif income >= 3000 and credit_score < 700:
    print("Loan denied: credit score too low.")
elif income < 3000 and credit_score >= 700:
    print("Loan denied: income too low.")
else:
    print("Loan denied: income and credit score too low.")
```

```python
# Example 10: study time recommendation
hours_available = float(input("How many hours do you have to study today? "))

if hours_available >= 5:
    print("Plan: Full study session with breaks.")
elif hours_available >= 2:
    print("Plan: Focus on the most important topics.")
elif hours_available > 0:
    print("Plan: Review notes quickly.")
else:
    print("Plan: No study time today.")
```
