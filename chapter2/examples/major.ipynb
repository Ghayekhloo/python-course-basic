## 1. What is `if / elif / else`?

In Python, we use conditional statements to make decisions.

### Basic structure
```python
if condition:
    # code
elif another_condition:
    # code
else:
    # code
```

Python checks conditions from top to bottom:
1. If the first condition is `True`, that block runs.
2. If not, Python checks the `elif` condition.
3. If none of the conditions are `True`, the `else` block runs.


````python
# Simple example
score = 82

if score >= 90:
    print("Excellent")
elif score >= 75:
    print("Good job")
else:
    print("Keep practicing")
````

## 4. Example: Advice by Major

Now we use `if / elif / else` to print different messages for different academic fields.


````python
major = "Marine ecology"

if major == "Supply chain management":
    print("Focus on logistics, inventory, and global operations.")
elif major == "Communication Design":
    print("Focus on visual identity, typography, and user-centered communication.")
elif major == "Marine ecology":
    print("Focus on ocean ecosystems, species interaction, and environmental monitoring.")
elif major == "Electronics Engineering":
    print("Focus on circuits, sensors, and embedded systems.")
else:
    print("This major has its own unique learning path.")
````

## 5. Multiple Mechanical Engineering Cases

Some students have similar majors.
We can use logical operators to group related fields.


````python
major = "Mechanical manufacturing management and aerospace innovation"

if major == "Mechanical Engineering" or major == "Mechanical engineer" or major == "mechanical engineer":
    print("This student studies machines, motion, energy, and design.")
elif major == "Mechanical manufacturing management and aerospace innovation":
    print("This student combines manufacturing, innovation, and aerospace systems.")
else:
    print("This is another field.")
````

## 6. Using `in` for Cleaner Conditions

Sometimes several majors belong to the same broad category.


````python
major = "Biodiversity"

life_science_majors = ["Marine ecology", "Biodiversity", "Vegetation science", "ecological Modeling"]
engineering_majors = ["Electronics Engineering", "Mechanical Engineering", "Mechanical engineer", "mechanical engineer"]

if major in life_science_majors:
    print("This major belongs to environmental or life sciences.")
elif major in engineering_majors:
    print("This major belongs to engineering.")
else:
    print("This major belongs to another category.")
````

## 7. GPA + Major + Status

Here we combine:
- strings
- float
- boolean
- logical operators
- conditional statements


````python
student_name = "Nora"
major = "Supply chain management"
gpa = 3.8
is_full_time = True

if major == "Supply chain management" and gpa >= 3.5 and is_full_time:
    print(f"{student_name} is an excellent candidate for logistics and operations internships.")
elif major == "Supply chain management" and gpa >= 3.0:
    print(f"{student_name} is doing well and should continue building analytical skills.")
else:
    print(f"{student_name} should focus on improving academic performance and practical experience.")
````

## 8. Nested `if` Example

A nested `if` means an `if` statement inside another `if` statement.


````python
major = "Electronics Engineering"
gpa = 3.7

if major == "Electronics Engineering":
    print("Major accepted.")

    if gpa >= 3.5:
        print("Eligible for advanced electronics lab work.")
    else:
        print("Build stronger fundamentals before advanced lab work.")
else:
    print("This example is only for Electronics Engineering students.")
````

## 11. Final Comprehensive Example

This example combines:
- strings
- numbers
- booleans
- comparison operators
- logical operators
- `if / elif / else`
- nested decision-making


````python
name = input("Student name: ")
major = input("Major: ")
gpa = float(input("GPA: "))
has_project = input("Completed project? (yes/no): ").lower() == "yes"
english_level = input("English level (high/medium/low): ").lower()

life_science_majors = ["Marine ecology", "Biodiversity", "Vegetation science", "ecological Modeling"]
engineering_majors = ["Electronics Engineering", "Mechanical Engineering", "Mechanical engineer", "mechanical engineer"]
design_business_majors = ["Communication Design", "Supply chain management", "Mechanical manufacturing management and aerospace innovation"]

if major in life_science_majors:
    if gpa >= 3.5 and has_project:
        print(f"{name} is well prepared for environmental research opportunities.")
    elif gpa >= 3.0:
        print(f"{name} has a good foundation in {major}.")
    else:
        print(f"{name} should strengthen research and data analysis skills.")
elif major in engineering_majors:
    if gpa >= 3.5 and english_level == "high":
        print(f"{name} is a strong candidate for technical projects and international collaboration.")
    elif gpa >= 3.0:
        print(f"{name} is progressing well in {major}.")
    else:
        print(f"{name} should improve both technical fundamentals and practice.")
elif major in design_business_majors:
    if has_project and english_level in ["high", "medium"]:
        print(f"{name} is ready for presentations, teamwork, and applied projects in {major}.")
    elif gpa >= 3.0:
        print(f"{name} has potential and should keep building professional skills.")
    else:
        print(f"{name} should improve both academic confidence and project experience.")
else:
    print(f"{name} is in a unique field. Build your strengths step by step.")
````
