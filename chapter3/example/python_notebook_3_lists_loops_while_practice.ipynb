## 1. What is a List?

A **list** is a collection of items in Python.

We use square brackets `[]` to create a list.

### Example
```python
majors = ["Marine ecology", "Mechanical Engineering", "Communication Design"]
```

A list can store:
- strings
- numbers
- booleans
- or mixed data types


````python
majors = [
    "Supply chain management",
    "Communication Design",
    "Marine ecology",
    "Electronics Engineering",
    "Biodiversity",
    "Mechanical engineer",
    "Vegetation science",
    "Mechanical Engineering",
    "ecological Modeling",
    "Mechanical manufacturing management and aerospace innovation"
]

print(majors)
print(type(majors))
print("Number of majors:", len(majors))
````

## 2. Accessing List Items

We can access items by index.

Remember:
- the first item has index `0`
- the second item has index `1`


````python
majors = [
    "Supply chain management",
    "Communication Design",
    "Marine ecology",
    "Electronics Engineering"
]

print(majors[0])
print(majors[1])
print(majors[2])
print(majors[-1])
````

## 3. Updating and Adding Items

We can change list items or add new ones.


````python
majors = ["Marine ecology", "Biodiversity", "Vegetation science"]

majors[1] = "Ecology and Biodiversity"
majors.append("ecological Modeling")

print(majors)
````

## 5. `for` Loop

A `for` loop repeats code for each item in a list.

### Structure
```python
for item in list_name:
    print(item)
```


````python
student_majors = [
    "Supply chain management",
    "Communication Design",
    "Marine ecology",
    "Electronics Engineering",
    "Biodiversity"
]

for major in student_majors:
    print(f"This student is studying {major}.")
````

## 6. `for` Loop with Conditions

Now we combine loops with `if / elif / else`.


````python
student_majors = [
    "Supply chain management",
    "Communication Design",
    "Marine ecology",
    "Electronics Engineering",
    "Biodiversity",
    "Mechanical Engineering"
]

for major in student_majors:
    if major == "Supply chain management":
        print(f"{major}: focus on logistics, planning, and operations.")
    elif major == "Communication Design":
        print(f"{major}: focus on visual storytelling and creative communication.")
    elif major == "Marine ecology":
        print(f"{major}: focus on marine ecosystems and environmental observation.")
    elif major == "Electronics Engineering":
        print(f"{major}: focus on circuits, sensors, and smart systems.")
    elif major == "Biodiversity":
        print(f"{major}: focus on species diversity and conservation.")
    else:
        print(f"{major}: this major has a special engineering pathway.")
````

## 7. Counting Specific Types of Majors

We can use a loop and a counter.


````python
student_majors = [
    "Marine ecology",
    "Biodiversity",
    "Vegetation science",
    "Mechanical Engineering",
    "Electronics Engineering",
    "ecological Modeling"
]

science_count = 0
engineering_count = 0

for major in student_majors:
    if major in ["Marine ecology", "Biodiversity", "Vegetation science", "ecological Modeling"]:
        science_count += 1
    else:
        engineering_count += 1

print("Science majors:", science_count)
print("Engineering majors:", engineering_count)
````

## 8. Working with Two Lists

We can connect names and majors using two lists.


````python
student_names = ["Lina", "Omar", "Sara", "Daniel", "Nora"]
student_majors = [
    "Marine ecology",
    "Mechanical Engineering",
    "Communication Design",
    "Electronics Engineering",
    "Biodiversity"
]

for i in range(len(student_names)):
    print(f"{student_names[i]} studies {student_majors[i]}.")
````

## 9. Loop + GPA + Major

This example uses lists, loops, conditions, and f-strings together.


````python
student_names = ["Lina", "Omar", "Sara", "Daniel", "Nora"]
student_majors = [
    "Marine ecology",
    "Mechanical Engineering",
    "Communication Design",
    "Electronics Engineering",
    "Biodiversity"
]
gpas = [3.8, 3.2, 3.9, 2.8, 3.5]

for i in range(len(student_names)):
    name = student_names[i]
    major = student_majors[i]
    gpa = gpas[i]

    if gpa >= 3.7:
        print(f"{name} from {major} is an outstanding student.")
    elif gpa >= 3.0:
        print(f"{name} from {major} is doing well.")
    else:
        print(f"{name} from {major} should practice more.")
````

## 11. Introduction to `while`

A `while` loop repeats code **while a condition is True**.

### Structure
```python
while condition:
    # code
```

A `while` loop is useful when we do not know in advance how many times we need to repeat something.


````python
count = 1

while count <= 5:
    print("Count:", count)
    count += 1
````

## 12. `while` Loop with Student Practice

Let us print a list of majors using a `while` loop.


````python
student_majors = [
    "Supply chain management",
    "Communication Design",
    "Marine ecology",
    "Electronics Engineering",
    "Biodiversity"
]

i = 0

while i < len(student_majors):
    print(f"Major {i+1}: {student_majors[i]}")
    i += 1
````

## 13. `while` + Condition Example

We can also use `while` with conditions about student scores.


````python
scores = [88, 74, 91, 69, 83]
i = 0

while i < len(scores):
    if scores[i] >= 85:
        print(f"Score {scores[i]}: Excellent")
    elif scores[i] >= 70:
        print(f"Score {scores[i]}: Good")
    else:
        print(f"Score {scores[i]}: Needs more practice")
    i += 1
````

## 14. Multiple Conditions Inside a Loop

This is an important practice section.


````python
student_names = ["Lina", "Omar", "Sara", "Daniel", "Nora"]
student_majors = [
    "Marine ecology",
    "Mechanical Engineering",
    "Communication Design",
    "Electronics Engineering",
    "Biodiversity"
]
gpas = [3.8, 3.2, 3.9, 2.8, 3.5]
has_project = [True, False, True, True, False]

for i in range(len(student_names)):
    name = student_names[i]
    major = student_majors[i]
    gpa = gpas[i]
    project = has_project[i]

    if gpa >= 3.7 and project:
        print(f"{name} in {major} is ready for advanced academic opportunities.")
    elif gpa >= 3.0 and not project:
        print(f"{name} in {major} has good grades but should complete a project.")
    elif gpa >= 3.0 and project:
        print(f"{name} in {major} is progressing well.")
    else:
        print(f"{name} in {major} should improve both practice and academic confidence.")
````

## 16. Final Comprehensive Example: Major Advisor

This final example combines:
- lists
- loops
- while
- conditions
- strings
- floats
- booleans
- user-style academic feedback


````python
student_names = ["Lina", "Omar", "Sara", "Daniel", "Nora", "Mila"]
student_majors = [
    "Marine ecology",
    "Mechanical Engineering",
    "Communication Design",
    "Electronics Engineering",
    "Biodiversity",
    "Supply chain management"
]
gpas = [3.8, 3.2, 3.9, 2.8, 3.5, 3.1]
english_levels = ["high", "medium", "high", "low", "medium", "high"]

i = 0

while i < len(student_names):
    name = student_names[i]
    major = student_majors[i]
    gpa = gpas[i]
    english = english_levels[i]

    if major == "Marine ecology":
        if gpa >= 3.5:
            print(f"{name}: strong profile for marine research and environmental fieldwork.")
        else:
            print(f"{name}: improve analysis and field practice in marine ecology.")

    elif major == "Mechanical Engineering":
        if gpa >= 3.0 and english in ["high", "medium"]:
            print(f"{name}: ready for engineering teamwork, reports, and technical growth.")
        else:
            print(f"{name}: strengthen both technical skills and communication.")

    elif major == "Communication Design":
        if gpa >= 3.7:
            print(f"{name}: excellent for portfolio development and creative leadership.")
        else:
            print(f"{name}: continue improving design practice and presentation skills.")

    elif major == "Electronics Engineering":
        if english == "high":
            print(f"{name}: good potential for international electronics projects.")
        else:
            print(f"{name}: improve both programming and technical English.")

    elif major == "Biodiversity":
        if gpa >= 3.5:
            print(f"{name}: ready for conservation and biodiversity research.")
        else:
            print(f"{name}: build stronger ecological observation and reporting skills.")

    else:
        if gpa >= 3.0:
            print(f"{name}: doing well in supply chain analysis and operations.")
        else:
            print(f"{name}: needs more practice in planning and analytical thinking.")

    i += 1
````
