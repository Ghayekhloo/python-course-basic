## 1. What is a Variable?
A variable is like a **labeled box**.
- The label = variable name
- The content = value

We create a variable with: `variable_name = value`

```python
# Variables store data
name = "Alessandro"
age = 25
height = 1.75
is_student = True

print("name:", name)
print("age:", age)
print("height:", height)
print("Is a student?", is_student)
```

## 2. Basic Data Types

| Type | Keyword | Example |
|------|---------|--------|
| Text | `str` | `"Hello"` |
| Whole number | `int` | `25` |
| Decimal number | `float` | `3.14` |
| True / False | `bool` | `True` |

```python
x = "Hello"
y = 20
c = 5.2
d = True

print(type(x))
print(type(y))
print(type(c))
print(type(d))
```

## 3. String (`str`)

```python
name = "Maria"
course = "Python Basic Level"
your_email = "maria123@gmail.com"

print(name)
print(your_email)
```

```python
name = "Omar"
last_name = "Ahmad"
print(f"My name is {name} {last_name}")

experiment = "DNA Analysis"
lab = "Biotech Lab"
print(f"We are doing {experiment} in the {lab}")
```

## 4. Integer (`int`)

```python
age = 25
grade = -5
print(age)
print(grade)
print("How old are you?", age, "years")
```

```python
a = 10
b = 5
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a ** b =", a ** b)
print("a % b =", a % b)
```

## 5. Float (`float`)

```python
price = 19.99
temperature = 21.8
speed = 75.5
print(f"The price is {price} euros and the speed is {speed} km/h.")
```

## 6. Boolean (`bool`)

> ⚠️ `True` and `False` must start with a **capital letter**.

```python
is_student = True
engine_running = True
cell_alive = False
print(is_student)
print("Engine running?", engine_running)
print("Cell alive?", cell_alive)
```

## 7. All 4 Types Together

```python
name = "Matteo"
age = 28
cell_size = 0.002
cell_active = True
print(f"My name is {name}, I am {age} years old.")
print(f"Cell size: {cell_size} mm. Active: {cell_active}")
```

## 8. Comparison Operators

| Operator | Meaning |
|----------|---------|
| `>` | greater than |
| `<` | less than |
| `==` | equal |
| `!=` | not equal |
| `>=` | greater or equal |
| `<=` | less or equal |

```python
age = 18
print(age > 20)
print(age < 20)
print(age == 18)
print(age != 22)
temperature = 15.9
print(temperature <= 30)
```

## 9. Logical Operators

| Operator | Meaning |
|----------|---------|
| `and` | both True |
| `or` | at least one True |
| `not` | reverse |

```python
is_student = True
has_id = False
print(is_student and has_id)
print(is_student or has_id)
print(not has_id)
```

## 10. Input from User

> ⚠️ `input()` always returns a **string**. Use `int()` or `float()` to convert.

```python
name = input("What's your name? ")
print(f"Hello {name}")
print(type(name))
```

```python
age = int(input("Enter your age: "))
height = float(input("Enter your height in meters: "))
print("Your age is", age, "years old.")
print("Your height is", height, "meters.")
print(type(age))
print(type(height))
```

## 11. Final Example

```python
animal_name = input("Enter name of animal: ")
weight = float(input("Enter animal weight (kg): "))
if weight > 100:
    level = "high"
elif weight > 50:
    level = "medium"
else:
    level = "low"
print(f"The weight level of {animal_name} is {level} (weight: {weight} kg).")
```
