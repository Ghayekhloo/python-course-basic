## 1. What is a Variable?
A variable stores data.

Example:

```python
name = 'Sarah'
age = 42
print(name, age)
```

## 2. Basic Data Types
- int
- float
- str
- bool


```python
x = 10
print(type(x))
y = 3.5
print(type(y))
text = 'Hello'
print(type(text))
flag = True
print(type(flag))
```

## 3. String
Strings are text data.

Examples from real student fields:

```python
majors = ['Supply chain management', 'Communication Design', 'Marine ecology', 'Electronics Engineering', 'Biodiversity', 'Mechanical engineer', 'Vegetation science', 'Mechanical Engineering', 'ecological Modeling', 'Mechanical manufacturing management and aerospace innovation']
for m in majors:
    print('Field:', m)
```

```python
# f-string example
name = 'Student'
field = 'Mechanical Engineering'
print(f'{name} studies {field}.')
```

## 4. Integer

```python
a = 5
b = 2
print(a + b, a * b, a ** b)
```

## 5. Float

```python
x = 5.5
y = 2.0
print(x / y, x + y)
```

## 6. Boolean

```python
is_student = True
print(is_student)
```

## 7. Combined Example

```python
name = 'Ali'
age = 21
gpa = 3.7
is_active = True
print(name, age, gpa, is_active)
```

## 8. Comparison Operators

```python
print(5 > 3)
print(5 == 5)
print(5 != 2)
```

## 9. Logical Operators

```python
print(True and False)
print(True or False)
print(not True)
```

## 10. User Input

```python
age = int(input('Enter your age: '))
print(age + 1)
```

## 11. Final Example

```python
name = input('Name: ')
field = input('Field: ')
age = int(input('Age: '))

if age > 18:
    print(f'{name} studies {field} and is an adult.')
else:
    print(f'{name} studies {field} and is young.')
```
